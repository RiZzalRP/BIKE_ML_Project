```python
df.Season=df.Season.fillna(df.Season.mode()[0])
df.Year=df.Year.fillna(df.Year.mode()[0])
df.Month=df.Month.fillna(df.Month.mode()[0])
df.Hour=df.Hour.fillna(df.Hour.mode()[0])
df.Holiday=df.Holiday.fillna(df.Holiday.mode()[0])
df.Weekend=df.Weekend.fillna(df.Weekend.mode()[0])
df['Working Day']=df['Working Day'].fillna(df['Working Day'].mode()[0])
df.Weather=df.Weather.fillna(df.Weather.mode()[0])
df.Temperature=df.Temperature.fillna(df.Temperature.mean())
df.aTemp=df.aTemp.fillna(df.aTemp.mean())
df.Humidity=df.Humidity.fillna(df.Humidity.mean())
df.WindSpeed=df.WindSpeed.fillna(df.WindSpeed.mean())
df.Casual=df.Casual.fillna(df.Casual.mode()[0])
df.Registered=df.Registered.fillna(df.Registered.mode()[0])
```

```python
# Find the limits
upper_limit=df.Hour.mean() + 3*df.Hour.std()
lower_limit=df.Hour.mean() - 3*df.Hour.std()
print('Upper_Bound :',upper_limit)
print('Lower_Bound :',lower_limit)
```

```python
# Find the Outlier
outlier=df.loc[(df.Hour > upper_limit) | (df.Hour < lower_limit)]
outlier
```

```python
# Drop the outlier
df=df[(df.Hour >=lower_limit) & (df.Hour <=upper_limit)]
df
```

```python
x=df.drop(['Count'],axis=1)
x.ndim
```

```python
y=df['Count']
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,HuberRegressor,ElasticNetCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,ExtraTreesRegressor
LR_model=LinearRegression()
R_model=Ridge()
HR_model=HuberRegressor()
E_model=ElasticNetCV()
KNN_model=KNeighborsRegressor()
DR_model=DecisionTreeRegressor()
RR_model=RandomForestRegressor()
GR_model=GradientBoostingRegressor()
ER_model=ExtraTreesRegressor()
models=[LR_model,R_model,HR_model,HR_model,E_model,KNN_model,DR_model,RR_model,GR_model,ER_model]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
for model in models:
  print('MODEL :',model)
  model.fit(x_train,y_train)
  y_pred=model.predict(abs(x_test))
  print('Y_PREDICT of',model,':',y_pred)
  print()
  print('.....Calculate Mean Squared Error.....')
  print('MSE_',model,':',mean_squared_error(y_test, y_pred))
  print()
  print('.....Calculate Root Mean Squared Error.....')
  print('RMSE_',model,':',np.sqrt(mean_squared_error(y_test, y_pred)))
  print()
  print('.....Calculate Mean Absolute Error.....')
  print('MAE_',model,':',mean_absolute_error(y_test, y_pred))
  print()
  print('.....Calculate R-squared.....')
  print('R2-Squre_',model,':',r2_score(y_test, y_pred))
  print()
  print('\n')
```

```python
from sklearn.model_selection import GridSearchCV
params={'n_estimators':[100,300,500],'max_depth':[1,3,5,7,10],'min_samples_leaf':[1,3,5,7,10],'criterion':["squared_error","absolute_error","friedman_mse"],'min_samples_split':[2,4,6,8,10],'min_samples_leaf':[1,3,5,7],'bootstrap':[True,False]}
rr_model=RandomForestRegressor()
gsc=GridSearchCV(rr_model,params,cv=10,scoring='accuracy')
gsc.fit(x_train,y_train)
print(gsc.best_score_)
print(gsc.best_params_)
```

```python
model = RandomForestRegressor(n_estimators=100,min_samples_leaf=1,max_depth=1)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
error=y_test - y_pred
fig,ax=plt.subplots()
ax.scatter(y_test,error)
ax.axhline(lw=3,color='black',linestyle='--')
ax.set_title('Error Graph')
ax.set_xlabel('Observed')
ax.set_ylabel('Error')
plt.show()
```

```python
#error analysis graph of this model
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='green', alpha=0.4)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='blue', linestyle='-.')
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid(True)
plt.show()
```
